This is to include region name, local authority name, domestic total emissions, domestic electricity emissions, domestic gas emissions, per capita emissions and year.

```python
New=CO2[['Region Name','LAD14NM','Domestic Total','F. Domestic Electricity','G. Domestic Gas','Per Capita Emissions (t)','Year']]
```

### Clean and process

Remove rows with missing values:

```python
New=New.dropna()
```

Rename local authority column:

```python
New=New.rename(columns={"LAD14NM":"Area Name"})
```

Convert year to an integer value:

```python
New['Year']=New['Year'].astype('int64')
```

Remove commas in numbers:

```python
New['Domestic Total']=New['Domestic Total'].str.replace(",","")
New['F. Domestic Electricity']=New['F. Domestic Electricity'].str.replace(",","")
New['G. Domestic Gas']=New['G. Domestic Gas'].str.replace(",","")
New['Per Capita Emissions (t)']=New['Per Capita Emissions (t)'].str.replace(",","")
```

Remove dashes in data:

```python
New=New.replace("-","")
```

Convert objects to numeric:

```python
New['Domestic Total']=pd.to_numeric(New['Domestic Total'])
New['F. Domestic Electricity']=pd.to_numeric(New['F. Domestic Electricity'])
New['G. Domestic Gas']=pd.to_numeric(New['G. Domestic Gas'])
New['Per Capita Emissions (t)']=pd.to_numeric(New['Per Capita Emissions (t)'])
```

Check datatypes:

```python
New.dtypes
```

Rename columns to remove letters:

```python
New=New.rename(columns={"F. Domestic Electricity":"Domestic Electricity","G. Domestic Gas":"Domestic Gas"})
```

Find 'unallocated' regions (these will be removed):

```python
New[(New['Region Name'].str.contains("Unallocated"))]
```

Remove entries relateing to Northern Ireland, Wales and Scotland:

```python
New=New[~New['Region Name'].str.contains("Northern Ireland")]
New=New[~New['Region Name'].str.contains("Scotland")]
New=New[~New['Region Name'].str.contains("Wales")]
```

Remove entries relating to area totals and unallocated regions:

```python
New=New[~New['Area Name'].str.contains("Total")]
New=New[~New['Area Name'].str.contains("Unallocated")]
New=New[~New['Region Name'].str.contains("Total")]
New=New[~New['Region Name'].str.contains("Unallocated")]
New
```

Count local authority areas:

#### Calculate % change from 2011 to 2014 in fuel poverty and domestic total emissions among these local authorities

Create new columns showing % change for fuel poverty and domestic total emissions:

```python
WorstDomFP_11_14['2011 to 2014 Fuel Poverty % Change']=(((WorstDomFP_11_14['2014 Fuel Poverty'])-(WorstDomFP_11_14['2011 Fuel Poverty']))/WorstDomFP_11_14['2011 Fuel Poverty'])*100
WorstDomFP_11_14['2011 to 2014 Domestic Total CO2 Emissions % Change']=(((WorstDomFP_11_14['2014 Domestic Total'])-(WorstDomFP_11_14['2011 Domestic Total']))/WorstDomFP_11_14['2011 Domestic Total'])*100
```

Round % change to one decimal place:

```python
WorstDomFP_11_14['2011 to 2014 Fuel Poverty % Change']=WorstDomFP_11_14['2011 to 2014 Fuel Poverty % Change'].round(1)
WorstDomFP_11_14['2011 to 2014 Domestic Total CO2 Emissions % Change']=WorstDomFP_11_14['2011 to 2014 Domestic Total CO2 Emissions % Change'].round(1)
```

Sort values by % change in fuel poverty:

```python
WorstDomFP_11_14sorted=WorstDomFP_11_14.sort_values('2011 to 2014 Fuel Poverty % Change')
WorstDomFP_11_14sorted
```

#### Plot % change in fuel poverty and domestic emissions among these local authorities

```python
WorstDomFP_11_14graph=WorstDomFP_11_14sorted[['2011 to 2014 Fuel Poverty % Change','2011 to 2014 Domestic Total CO2 Emissions % Change']].plot.bar(figsize=(10,5),title='Fuel Poverty and Domestic Total CO2 Emissions of Worst LAs: 2011 to 2014 % Change')
WorstDomFP_11_14graph.set_ylabel('2011 to 2014 % Change')
WorstDomFP_11_14graph.set_xlabel('Local Authority')
```
